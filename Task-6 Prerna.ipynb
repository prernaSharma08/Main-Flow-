```python
director_counts = pd.crosstab(index=data['director'], columns='count')
director_counts_sorted = director_counts.sort_values(by='count', ascending=False)
top_three_directors = director_counts_sorted.head(3)
print("Top three directors with the most content:")
print(top_three_directors)
```

```python
director_counts = pd.crosstab(index=data['director'], columns='count')
director_counts_sorted = director_counts.sort_values(by='count', ascending=False)
top_three_directors = director_counts_sorted.head(3)
print("Top three directors with the most content:")
print(top_three_directors)
```

```python
top_three_directors = director_counts_sorted.head(3)
top_three_directors = pd.DataFrame({'count': [10, 20, 30]}, index=['Director A', 'Director B', 'Director C'])
plt.figure(figsize=(8, 4))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
sns.barplot(x=top_three_directors.index, y='count', data=top_three_directors, palette=colors)
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.title('Top Three Directors with Most Content')
plt.show()
```

```python
country_counts = data['country'].value_counts()
top_country = country_counts.idxmax()
highest_count = country_counts.max()
print("Country with the highest number of movies and TV shows combined:")
print("Country:", top_country)
print("Number of titles:", highest_count)
```

```python
country_counts = data['country'].value_counts()
top_countries = country_counts.head(5)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.title('Top 5 Countries with the Highest Number of Movies and TV Shows Combined')
plt.tight_layout()
plt.show()
```
